```python
import numpy as np
import matplotlib.pyplot as plt

# Definisi fungsi
def f(x):
    return -x**2 + 4*x + 5

# Nilai x untuk plotting
x = np.linspace(-1, 5, 100)
y = f(x)

# Titik maksimum
x_max = 2
y_max = f(x_max)

# Plot grafik
plt.figure(figsize=(8, 5))
plt.plot(x, y, label=r'$f(x) = -x^2 + 4x + 5$', color='blue')
plt.scatter(x_max, y_max, color='red', zorder=3, label=f'Maksimum di ({x_max}, {y_max})')
plt.axvline(x_max, linestyle='--', color='gray', alpha=0.7)
plt.axhline(y_max, linestyle='--', color='gray', alpha=0.7)

# Label dan tampilan
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Grafik Fungsi $f(x) = -x^2 + 4x + 5$')
plt.legend()
plt.grid()
plt.show()
```

# Optimasi Lagrange

```python
# Grid untuk plotting
x_vals = np.linspace(0, 10, 100)
y_vals = 10 - x_vals  # Kendala x + y = 10

# Fungsi tujuan (kontur)
X, Y = np.meshgrid(x_vals, x_vals)
Z = X * Y

# Plot
plt.figure(figsize=(8, 6))
contour = plt.contour(X, Y, Z, levels=20, cmap='coolwarm')
plt.clabel(contour, inline=True, fontsize=8)
plt.plot(x_vals, y_vals, 'k--', label=r'Kendala: $x + y = 10$')

# Titik optimal
plt.scatter(5, 5, color='red', label=r'Maksimum di $(5,5)$', zorder=3)

# Label dan tampilan
plt.xlabel('x')
plt.ylabel('y')
plt.title('Optimasi dengan Metode Lagrange')
plt.legend()
plt.grid()
plt.show()
```

# Gradient Descent

```python
# Definisi fungsi dan gradien
def f(x):
    return 2*x**2 - 48*x + 9

def grad_f(x):
    return 4*x - 48

# Inisialisasi parameter
x_t = 6000  # Titik awal
alpha = 0.1  # Learning rate
# iterations = 15  # Jumlah iterasi
history = [x_t]  # Menyimpan nilai x setiap iterasi

# Gradient Descent Iterasi
i = 0
while abs(grad_f(x_t)) > 1e-9:
    x_t = x_t - alpha * grad_f(x_t)
    history.append(x_t)
    print(f'{i}: x = {x_t:.10f}, f(x) = {f(x_t):.10f}, grad_f(x) = {grad_f(x_t):.10f}')
    i += 1

# Plot fungsi
x_vals = np.linspace(0, 8, 100)
y_vals = f(x_vals)

plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, label=r'$f(x) = x^2 - 4x + 4$', color='blue')
plt.scatter(history, [f(x) for x in history], color='red', label='Iterasi Gradient Descent', zorder=3)
plt.plot(history, [f(x) for x in history], '--r', alpha=0.7)

# Minimum global
x_min = 2
y_min = f(x_min)
plt.scatter(x_min, y_min, color='green', s=100, label='Minimum di $(2,0)$')

# Label dan tampilan
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent untuk $f(x) = x^2 - 4x + 4$')
plt.legend()
plt.grid()
plt.show()
```

# Newton Method

```python
# Definisi fungsi dan gradien
def f(x):
    return 2*x**2 - 48*x + 9

def grad_f(x):
    return 4*x - 48

f2 = 4
# Inisialisasi parameter
x_t = 4412341  # Titik awal
history = [x_t]  # Menyimpan nilai x setiap iterasi

# Newton Iterasi
i = 0
print(f'{i}: x = {x_t:.10f}, f(x) = {f(x_t):.10f}, grad_f(x) = {(grad_f(x_t)/f2):.10f}')
while abs(grad_f(x_t)) > 1e-9:
    x_t = x_t - grad_f(x_t)/f2
    history.append(x_t)
    print(f'{i}: x = {x_t:.10f}, f(x) = {f(x_t):.10f}, grad_f(x) = {grad_f(x_t):.10f}')
    i += 1

# Plot fungsi
x_vals = np.linspace(0, 8, 100)
y_vals = f(x_vals)

plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, label=r'$f(x) = x^2 - 4x + 4$', color='blue')
plt.scatter(history, [f(x) for x in history], color='red', label='Iterasi Gradient Descent', zorder=3)
plt.plot(history, [f(x) for x in history], '--r', alpha=0.7)

# Minimum global
x_min = 2
y_min = f(x_min)
plt.scatter(x_min, y_min, color='green', s=100, label='Minimum di $(2,0)$')

# Label dan tampilan
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent untuk $f(x) = x^2 - 4x + 4$')
plt.legend()
plt.grid()
plt.show()
```

# Regresi Linear

```python
import numpy as np
import matplotlib.pyplot as plt

# Data sampel (X, y)
X = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 2, 2, 3, 3, 6])  # Linear dengan w=2, b=0

# Gradient Descent Parameters
alpha = 0.01  # Learning rate
iterations = 1000
m = len(X)
w, b = 0, 0  # Inisialisasi parameter

# Gradient Descent
for _ in range(iterations):
    y_pred = w * X + b
    dw = (-2/m) * np.sum(X * (y - y_pred))
    db = (-2/m) * np.sum(y - y_pred)
    w -= alpha * dw
    b -= alpha * db

print(f"Gradient Descent: w = {w:.4f}, b = {b:.4f}")


# Plot hasil regresi
plt.scatter(X, y, color='blue', label='Data Asli')
plt.plot(X, w * X + b, color='red', label='Gradient Descent')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Optimasi Regresi Linear: Gradient Descent vs Newton')
plt.show()
```
